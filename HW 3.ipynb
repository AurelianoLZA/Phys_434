### Part A :

from the Wiki webpage, we found that if we have 2 exactly same exponential function, which means we have the same value of $\lambda$, the resulting distrubution is still exponential distribution with different value of $\lambda$

```python
x = np.linspace(1+stats.expon.ppf(0.01),stats.expon.ppf(0.99),1000)
pdf = stats.expon.pdf(x,1)
```

```python
fig,ax = plt.subplots()
ax.plot(x,pdf,'r--',label="exponential distrubution")
ax.set_xlabel("$x$")
ax.set_ylabel("exponential pdf")
ax.legend()
```

```python
def convolv(A,old,count):
    if(count == 1):
        return old
    else:
        temp = np.convolve(A,old)
        return convolv(A,temp,count-1)
```

```python
new_pdf = convolv(pdf,pdf,2)
new_x = np.linspace(1,len(new_pdf),len(new_pdf))
```

```python
fig1,ax1 = plt.subplots()
ax1.plot(new_x,new_pdf,'r--',label="summation of 2 exponential pdf")
ax1.set_xlabel("$x$")
ax1.set_ylabel("new pdf")
ax1.legend()
```

#### Conclusion : From both analytically and graphically, we can see that the sum of two exponential pdf is a Gamma distribution

```python

sigma_x = 6.
sigma_y = 2.

x = np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), 1000)
y = np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), 1000)

x, y = np.meshgrid(x, y)
z = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x**2/(2*sigma_x**2)
     + y**2/(2*sigma_y**2))))

plt.contourf(x, y, z, cmap='Blues')
plt.title("Two Dimensional Gaussian distribuition")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.colorbar()
```
