#### Question 2

Based on the plots we found from above, we can see there exist some correlations between features, for instance:

From the plot between feature "ee2" and "d2", we can see a very clear trend that the value of "ee2" has inverse correlation with the value of "d2" and such trend persists in both signal dataset and background dataset. 

By carefully examining the source code, we can also find that there is correlation between the feature of "ee2", "ee3" and "d2". However, since their correlation involves more variables than just 2 where we can simply plot them out and see, we need to verify it as following:

```python
fig, ax1 = plt.subplots(1, 1)
start = 0
end = 100000
ax1.vlines(bcg['ee3']/(bcg['ee2'])**3, 0, bcg['d2'], colors='b', linestyles='-', lw=5, alpha = 0.3)
ax1.vlines(higgs['ee3']/(higgs['ee2'])**3, 0, higgs['d2'], colors='r', linestyles='-', lw=5, alpha = 0.3)
plt.show()
```

From the plot above, we can clearly see a linear pattern that exists between these 3 varaibles, and the slope is 1.
