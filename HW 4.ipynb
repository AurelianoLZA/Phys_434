```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import stats
```

### Problem 0

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

we can plot the messy data as below

```python
fig0,ax0 = plt.subplots()
ax0.scatter(t,bkgd,c = 'b',marker='o',linewidth = 0.1,label="messy scatter plot of all data")
ax0.set_xlabel("$t$")
ax0.set_ylabel("background")
ax0.legend()
```

As shown above, scatter plotting all data points are very un-informative. Instead, we can plot a subset of them. 

### Problem 1

#### Problem 1 a

```python
fig1,ax1 = plt.subplots()
ax1.scatter(t[:1000],bkgd[:1000],c = 'r',marker='*',label= "first 1000 points" )
ax1.set_xlabel("$t$")
ax1.set_ylabel("background")
ax1.legend()
```

#### Problem 1b

```python
fig2,ax2 = plt.subplots()
ax2.scatter(t[500:2500],bkgd[500:2500],c = 'r',marker='*',label="2000 points between 500 and 2500")
ax2.set_xlabel("$t$")
ax2.set_ylabel("background")
ax2.legend()
```

#### Porblem 1c : 

We can use the Python built-in slice function to "slice" the array, and then filter out evert 100th data

```python
s = slice(0,400,100)
fig3,ax3 = plt.subplots()
ax3.scatter(t[s],bkgd[s],c='r',marker='*',label="Every 100th points")
ax3.set_xlabel("$x$")
ax3.set_ylabel("background")
ax3.legend()
```

#### Problem 2 a

```python
fig4,ax4 = plt.subplots()
ax4.hist2d(t[30000:70000],bkgd[30000:70000],bins = 150, density =True,cmap = "plasma")
```

#### Problem 2b

#### The color in the plot above represents that for each bin we choose, how many points are in that bin; In other  words, it indicates how dense the plot is

### Problem 3

#### Question 3 a

```python
t = np.mod(np.linspace(0,np.pi*100,int(1e5)),np.pi*2)
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

```python
fig5,ax5 = plt.subplots()
ax5.scatter(t,bkgd,c='r',marker = "*",linewidth = 0.0001,label="data points after mod with $2\pi$")
ax5.set_xlabel("$x$")
ax5.set_ylabel("background")
ax5.legend()
```

#### Question 3 b

```python
fig6,ax6 = plt.subplots()
ax6.hist2d(t[30000:70000],bkgd[30000:70000],bins = 150, density =True,cmap="plasma")
```

#### From the plots above, we can see that the sinudiol pattern becomes to be much clearer after mod with 2$\pi$
